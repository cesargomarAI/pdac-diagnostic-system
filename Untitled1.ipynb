```python
import joblib
import numpy as np
from flask import Flask, request, jsonify

# Cargamos el modelo que tanto nos gusta
modelo = joblib.load('modelo_vinos.joblib')

app = Flask(__name__)

@app.route('/predict', methods=['POST'])
def predict():
    # Recibimos los datos del vino
    data = request.get_json(force=True)
    # Convertimos a formato que entiende el modelo
    prediction = modelo.predict([np.array(data['features'])])
    
    return jsonify({
        'vinedo_tipo': int(prediction[0]),
        'mensaje': 'Predicción realizada con éxito'
    })

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
